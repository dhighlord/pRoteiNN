```
import pandas as pd
import numpy as np

url_train='https://raw.githubusercontent.com/sak1b0/proteiNN/master/train_formatted.csv'
url_test='https://raw.githubusercontent.com/sak1b0/proteiNN/master/test_formatted.csv'

df_train=pd.read_csv(url_train,header=None)
df_test=pd.read_csv(url_test,header=None)


df_train=np.asarray(df_train)
df_test=np.asarray(df_test)

X_train = df_train[:, 0]
y_train = df_train[:, 1]

X_test = df_test[:, 0]
y_test = df_test[:, 1]

def debug_me():
  print(df_train.shape)
  print(X_train.shape)
  print(y_train.shape)

  print(df_test.shape)
  print(X_test.shape)
  print(y_test.shape)

debug_me()

#print(len(X_train[13][0]))

max_length_train=0
dumm='shovon'
for item in range (len(X_train)):
  if(len(X_train[item])>len(dumm)):
    dumm=X_train[item]
  max_length_train=max(max_length_train,len(X_train[item]))
  
  
print('Maximum length of train: ',max_length_train)
print(dumm)

max_length_test=0
for item in range (len(X_test)):
  max_length_test=max(max_length_test,len(X_test[item]))
  
print('Maximum length of test: ',max_length_test)


#print(X_train[0])
#print(X_train[0][0].replace('H','1'))
s=X_train[0]
#no. of unique letters (there goes my label encoding)
len(list(set(s)))

data = ''

for item in range (len(X_train)):
  X_train[item] = X_train[item]+'Z'*(max_length_train-len(X_train[item]))
  data = data + X_train[item]

for item in range (len(X_train)):
  y_train[item] = y_train[item]+'Z'*(max_length_train-len(y_train[item]))
  data = data + y_train[item]
  
print(X_train[0])


print(y_train[0])
```
