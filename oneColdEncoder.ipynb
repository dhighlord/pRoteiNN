```
import numpy as np

def OneColdEncoder(x):
  
  ret = list()
  
  encode = [[0 for i in range(26)] for j in range(26)]
  
  for i in range(26):
    encode[i][i] = 1
  
  ret = encode[ord(x)-65]
    
  return ret

a = np.array(['XYZ', 'DEF'])

s = list(a)

for item in range(len(s)):
  s[item]=list(s[item])
  
print(s)

k = []

for i in range(len(s)):
  t = []
  for item in range(len(s[i])):
    t.insert(2000, OneColdEncoder(s[i][item]))
    s[i][item]=ord(s[i][item])-64
  k.append(t)

print("S -> ", s)

print("K -> ", k)
print(k[0])
print(k[1])

a = np.array(k).reshape(2,3,26,1)

display(a)
```
